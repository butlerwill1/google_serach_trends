```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
```

## Initialize a connection to Google Trends

```python
pytrend = TrendReq()

keywords = ["Football", "Rugby", "Tennis"]
```

## Build payload to fetch data

```python
pytrend.build_payload(kw_list=keywords, timeframe='today 3-m')
```

### Retrieve data

```python
data = pytrend.interest_over_time()
```

#### Check and display the head of the dataframe

```python
print(data.head())
```

```python
plt.figure(figsize=(10, 5))

# Plot each keyword
for keyword in keywords:
    plt.plot(data.index, data[keyword], label=keyword, linewidth=2)

plt.title('Google Search Trends Over the Last 3 Months', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Search Interest', fontsize=12)
plt.legend(title='Keywords', title_fontsize='13', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
